```python
import os
import shutil
from pyspark.sql import SparkSession

# ── Criar SparkSession (SEM configs de committer) ──
spark = SparkSession.builder \
    .appName("Spark + Alarik") \
    \
    .config("spark.hadoop.fs.s3a.endpoint", "http://alarik:8080") \
    .config("spark.hadoop.fs.s3a.access.key", "DANLQMEACSNA6BQMHSPV") \
    .config("spark.hadoop.fs.s3a.secret.key", "iCj2+E7qMODIpuw7xGrxHCEGlpRbVPjz7AWVW06g") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false") \
    .config("spark.hadoop.fs.s3a.connection.timeout", "15000") \
    .config("spark.hadoop.fs.s3a.attempts.maximum", "5") \
    .config("spark.hadoop.fs.s3a.aws.credentials.provider",
            "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider") \
    .config("spark.driver.extraJavaOptions",
            "-Daws.java.v1.disableDeprecationAnnouncement=true") \
    .getOrCreate()

print("SparkSession criada com sucesso!")

# ── Criar DataFrame ──
df = spark.createDataFrame([
    (1, "Alice", 30),
    (2, "Bob", 25),
    (3, "Charlie", 35)
], ["id", "nome", "idade"])

# ── Escrever LOCAL primeiro (filesystem normal, sem S3) ──
local_path = "/tmp/parquet_upload"
if os.path.exists(local_path):
    shutil.rmtree(local_path)

df.coalesce(1).write.mode("overwrite").parquet(f"file://{local_path}")
print(f"Dados escritos localmente em {local_path}")

# ── Listar arquivos gerados ──
arquivos = []
for root, dirs, files in os.walk(local_path):
    for f in files:
        arquivos.append(os.path.join(root, f))
print(f"Arquivos gerados: {arquivos}")

# ── Copiar para S3 (Alarik) usando Hadoop FileSystem API ──
from org.apache.hadoop.fs import FileSystem
from org.apache.hadoop.fs import Path
from java.net import URI

fs = FileSystem.get(URI("s3a://bronze/"), spark._jsc.hadoopConfiguration())
s3_bucket = "bronze"
s3_prefix = "teste-upload"

for local_file in arquivos:
    filename = os.path.basename(local_file)
    s3_dest = Path(f"s3a://{s3_bucket}/{s3_prefix}/{filename}")
    fs.copyFromLocalFile(Path(local_file), s3_dest)
    print(f"Upload: {filename} -> s3a://{s3_bucket}/{s3_prefix}/{filename}")

print("\nUpload completo!")

# ── Ler de volta do S3 ──
df_lido = spark.read.parquet(f"s3a://{s3_bucket}/{s3_prefix}/")
df_lido.show()
```

```python
import boto3
import pandas as pd
import io

# -----------------------------
# Configurações S3
# -----------------------------
source_bucket = "bucket-origem"
target_bucket = "bucket-destino"
prefixo_destino = "meu-prefixo"  # "pasta virtual" no bucket

s3 = boto3.client(
    "s3",
    endpoint_url="http://alarik:8080",
    aws_access_key_id="DANLQMEACSNA6BQMHSPV",
    aws_secret_access_key="iCj2+E7qMODIpuw7xGrxHCEGlpRbVPjz7AWVW06g",
    region_name="us-east-1",
)

# -----------------------------
# 1. Listar arquivos no bucket origem
# -----------------------------
response = s3.list_objects_v2(Bucket=source_bucket)
parquet_files = [obj['Key'] for obj in response.get('Contents', []) if obj['Key'].endswith('.parquet')]

# -----------------------------
# 2. Ler cada arquivo e gravar no bucket destino com prefixo
# -----------------------------
for key in parquet_files:
    # Ler arquivo do S3 em memória
    obj = s3.get_object(Bucket=source_bucket, Key=key)
    df = pd.read_parquet(obj['Body'])
    
    # --- Transformação de exemplo ---
    df['nova_coluna'] = 'exemplo'

    # Definir o novo Key com prefixo
    filename = key.split('/')[-1]  # só o nome do arquivo
    new_key = f"{prefixo_destino}/{filename}"

    # Gravar no bucket destino usando buffer em memória
    buffer = io.BytesIO()
    df.to_parquet(buffer, index=False)
    buffer.seek(0)  # voltar pro início do buffer

    s3.put_object(Bucket=target_bucket, Key=new_key, Body=buffer)
    print(f"Arquivo gravado: s3://{target_bucket}/{new_key}")
```
