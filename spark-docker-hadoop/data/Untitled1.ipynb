```python
from pyspark.sql import Row

# Criar um DataFrame de exemplo
dados = [
    Row(id=1, produto="Notebook", valor=3500.0, regiao="SP"),
    Row(id=2, produto="Monitor",  valor=1200.0, regiao="RJ"),
    Row(id=3, produto="Teclado",  valor=350.0,  regiao="MG"),
]
df = spark.createDataFrame(dados)

bucket = "s3a://meu-bucket"

# Salvar em Parquet (recomendado)
df.write \
  .mode("overwrite") \
  .format("parquet") \
  .save(f"{bucket}/vendas/parquet/")

# Salvar em CSV
df.write \
  .mode("overwrite") \
  .format("csv") \
  .option("header", "true") \
  .save(f"{bucket}/vendas/csv/")

# Salvar particionado por região
df.write \
  .mode("overwrite") \
  .format("parquet") \
  .partitionBy("regiao") \
  .save(f"{bucket}/vendas/particionado/")
```
