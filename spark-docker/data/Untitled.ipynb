```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("teste-hdfs") \
    .config("spark.hadoop.dfs.client.use.datanode.hostname", "true") \
    .getOrCreate()
```

```python
spark = SparkSession.builder \
    .appName("teste-hdfs") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://hdfs-namenode:8020") \
    .getOrCreate()
```

```python
df = spark.createDataFrame([
    ("Alice", 30, "SP"),
    ("Bob", 25, "RJ"),
    ("Carol", 35, "MG"),
    ("David", 28, "SP"),
    ("Eva", 32, "RJ")
], ["nome", "idade", "estado"])
```

```python
df.show()
```

```python
df.write.mode("overwrite").parquet("hdfs://hdfs-namenode:8020/test/ui")
```

```python
df2 = spark.read.parquet("hdfs://hdfs-namenode:8020/dados/pessoas")
```

```python
df2.show()
```

```python
df.write.mode("overwrite").option('compression','gzip').parquet("hdfs://hdfs-namenode:8020/dados/pessoas")
```
